##Tratamento de dados

### Duplicados

```python
df_vendas.drop_duplicates(inplace=True)
```

```python
df_dez.drop_duplicates(inplace=True)
```

###Transformando os tipos de dados


```python
df_vendas.columns = list(i.strip().lower().replace(' ','_') for i in df_vendas.columns)
```

```python
df_dez.columns = list(i.strip().lower().replace(' ','_') for i in df_dez.columns)
```

```python
df_gerentes.columns = list(i.strip().lower().replace(' ','_') for i in df_gerentes.columns)
```

```python
df_vendas['valor_unitário'] = df_vendas['valor_unitário'].str.replace(r'R\$', '', regex=True)
```

```python
df_vendas['valor_unitário'] = df_vendas['valor_unitário'].astype(float)
```

##Concatenando os DataFrame (novas linhas)

```python
df_vendas_total = pd.concat([df_vendas,df_dez], axis=0, join='inner')
```

##Mesclando os dados (novas colunas)

```python
df_final = pd.merge(left=df_vendas_total, right=df_gerentes, on='id_loja',how='inner')
```

```python
df_final.info()
```

```python
df_final.drop(columns='código_venda', inplace=True)
```

```python
df_final['valor_final'].fillna(df_final['quantidade']*df_final['valor_unitário'], inplace=True)
```

1. Qual foi o faturamento analisado?


```python
faturamento = df_final['valor_final'].sum()
print(f'Faturamento:  R${faturamento:,.2f}')
```

2. Qual loja teve o maior faturamento?

```python
df_lj = df_final.groupby('id_loja')['valor_final'].sum().reset_index().sort_values(by='valor_final', ascending=False)
df_lj.head(1)
```

3. Qual produto foi o mais vendido em quantidade?


```python
df_qtd_prod = df_final.groupby('produto')['quantidade'].sum().reset_index().sort_values(by='quantidade', ascending=False)
df_qtd_prod.head(1)
```

4. Qual produto foi o menos vendido em quantidade?


```python
df_qtd_prod.tail(1)
```

5. Qual produto gerou a maior receita?

```python
df_receita_prod = df_final.groupby('produto')['valor_final'].sum().reset_index().sort_values(by='valor_final', ascending=False)

df_receita_prod.head(1)
```

6. Qual foi o mês com maior faturamento?

```python
df_final['mes'] = df_final['data'].dt.month
```

```python
df_fat_mes = df_final.groupby('mes')['valor_final'].sum().reset_index().sort_values(by='valor_final', ascending=False)

df_fat_mes['mes'] = df_fat_mes['mes'].replace({
    1:'Janeiro',
    2:'Fevereiro',
    3:'Março',
    4:'Abril',
    5:'Maio',
    6:'Junho',
    7:'Julho',
    8:'Agosto',
    9:'Setembro',
    10:'Outubro',
    11:'Novembro',
    12:'Dezembro'
})
df_fat_mes.head(1)
```

7. Quais foram os produtos mais vendidos por cada gerente?

```python
gerente = df_final.groupby(['gerente', 'produto'])['quantidade'].sum().reset_index()

idx = gerente.groupby('gerente')['quantidade'].idxmax()

df_top5 = gerente.loc[idx].sort_values(by='quantidade', ascending=False)

df_top5.head()
```
