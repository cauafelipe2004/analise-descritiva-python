##Passo 1: Carregamento e Exploração Inicial

```python
import pandas as pd
```

```python
df_vendas = pd.read_excel('Vendas.xlsx')
```

```python
df_vendas_dez = pd.read_excel('Vendas - Dez.xlsx')
```

```python
df_gerentes = pd.read_excel('Gerentes.xlsx')
```

```python
colunas = df_vendas.columns
colunas = list(i.strip().replace(' ', '_').lower() for i in colunas)
print(colunas)
```

```python
df_vendas.columns = colunas
```

##Passo 2: Consolidação e Limpeza dos Dados

```python
df_vendas['Valor Unitário'] = df_vendas['Valor Unitário'].str.replace('R$', '').astype(float)
```

```python
df_vendas_dez['Valor Final'] = df_vendas['Valor Final'].astype(float)
```

```python
df_unificada = pd.merge(df_vendas, df_vendas_dez, how='outer')
display(df_unificada)
```

```python
df_unificada['Data'] = pd.to_datetime(df_unificada['Data'], format= '%d/%m/%Y')
```

```python
display(df_unificada)
```

```python
df_unificada.loc[df_unificada['Valor Final'].isna(), 'Valor Final'] = df_unificada['Quantidade'] * df_unificada['Valor Unitário']
```

```python
display(df_unificada)
```

##Passo 3: Análise Descritiva e Perguntas de Negócio

```python
faturamento_total = df_unificada['Valor Final'].sum()
print(f'O faturamento total no período analisado foi de: R$ {faturamento_total:,.2f}')
```

```python
##Qual loja teve o maior faturamento?

loja_maior_faturamento = df_unificada.groupby('ID Loja')['Valor Final'].sum().sort_values(ascending=False)
display(loja_maior_faturamento)
```

```python
##Qual produto foi o mais vendido em quantidade?

prodt_mais_vendido = df_unificada.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
display(prodt_mais_vendido.head(1))
```

```python
##Qual produto foi o menos vendido em quantidade?

prodt_menos_vendido = df_unificada.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True)
display(prodt_menos_vendido.head(1))
```

```python
## Qual produto gerou a maior receita (somatório do Valor Final)?

prodt_maior_receita = df_unificada.groupby('Produto')['Valor Final'].sum().sort_values(ascending=False)
display(prodt_maior_receita.head(1))
```

##Passo 4: Mesclagem (Merge) e Análise Gerencial

```python
##Juntar o DataFrame consolidado de vendas com a base de gerentes, relacionando
cada venda ao gerente responsável.

df_unificada_gerencial = pd.merge(df_unificada, df_gerentes, how='outer')
display(df_unificada_gerencial)
```

```python
##Criar uma tabela com o total de vendas (em Valor Final) por gerente.

df_gerencial_vendas = df_unificada_gerencial.groupby(['Gerente'])[['Valor Final']].sum()
display(df_gerencial_vendas)
```

```python
##Identificar o gerente com maior volume de vendas em faturamento.

gerente_mais_vendas = df_gerencial_vendas.groupby('Gerente')['Valor Final'].sum().sort_values(ascending=False)
display(gerente_mais_vendas.head(1))
```

##Passo 5: Novos Problemas de Negócio

```python
##Qual foi o mês com maior faturamento?

df_unificada['Data'] = pd.to_datetime(df_unificada['Data'])
df_unificada['Mes'] = df_unificada['Data'].dt.to_period('M')
faturamento_mes = df_unificada.groupby('Mes')['Valor Final'].sum().sort_values(ascending=False)
display(faturamento_mes.head(1))
```

```python
## Quais foram os produtos mais vendidos por cada gerente?
idx = df_unificada_gerencial.groupby('Gerente')['Quantidade'].idxmax()
top_prod = df_unificada_gerencial.loc[idx].sort_values(by='Quantidade', ascending=False)

print(top_prod[['Gerente', 'Produto', 'Quantidade']])
```
